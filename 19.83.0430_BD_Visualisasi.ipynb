```python
# Import Library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Mempersiapkan Data
menu = ('Lava Cake', 'Mousse', 'Chocolate Cake', 'Ice Cream', 'Truffles', 'Brownies', 'Chocolate Chip Cookie', 
            'Chocolate Pudding', 'Souffle', 'Chocolate Cheesecake', 'Chocolate Chips', 'Fudge', 'Mochi')
jumlah = (14, 5, 12, 19, 6, 8, 12, 9, 10, 17, 2, 9, 13)

# Menu menjadi Data Frame
df = pd.DataFrame({
    'Menu':menu,
    'Jumlah':jumlah,
})

# Urutkan berdasarkan Jumlah
df.sort_values(by='Jumlah', inplace=True)

# Atur warna chart
x_coords = np.arange(len(df))
colors = ['#0000FF' for _ in range(len(df))]
colors[-13] = '#FF0000'
colors[-12] = '#FF0000'
colors[-11] = '#FF0000'


# Visualisasi bar chart
plt.bar(x_coords, df['Jumlah'], tick_label=df['Menu'], color=colors)
plt.xticks(rotation=90)
plt.ylabel('Qty Sold')
plt.title('Data Penjualan Desert')
plt.show()
```

Diliat dari presentasi penjualan diatas yang kita dapatkan ada 3 menu jualan dengan populalitas paling kecil yaitu Chocolate Chips, Mousse, Truffles oleh sebab itu maka 3 menu tersebut boleh di hapus dari 3 menu penjualan
