#Tuple
Tuple dalam Python adalah stuktur data yang digunakan untuk menyimpan sekumpulan data. Bersifat immutable, artinya isi tuple tidak bisa kita ubah dan hapus. Namun, dapat diisi dengan berbagai macam nilai dan objek. 

Source : https://www.w3schools.com/python/python_tuples.asp

```python
#Contoh Tuple 1
Bulan = ('Januari','Februari','Maret','April','Mei','Juni','Juli','Agustus','September','Oktober','November','Desember')
print("Jumlah bulan di satu tahun: %d" % len(Bulan))
```

#Set
Set adalah tipe data kolektif yang digunakan untuk menyimpan banyak nilai dalam satu variabel. Dengan beberapa detail khusus yaitu tidak ada duplikasi, bersifat unordered, nilai yang sudah dimasukkan bersifat unchangeable.

Source : https://www.w3schools.com/python/python_sets.asp , https://www.malasngoding.com/operasi-tipe-data-set/ 

```python
#Contoh Set 1
Contoh_Dosen_Idaman = {'Pak Anggit', 'Bu Senie', 'Pak Wahid', 'Bu Rina', 'Pak Jeki', 'Bu Rini'}
print(Contoh_Dosen_Idaman)
```

Tipe data kolektif ini sangat unik, python menyediakan kepada kita berbagai fungsi keanggotaan yang berguna untuk pengolahan data. 
Di antaranya : fungsi union (gabungan), serta fungsi intersection (irisan).

```python
#Contoh Set 2
ChelseaFC = {'Kai Havertz', 'Hakim Ziyech', 'Mason Mount', 'Ben Chilwell','Romelu Lukaku','Reece James','Thiago Silva'}
England = {'Harry Kane','Mason Mount','Jadon Sancho','Ben Chilwell','Reece James','Phil Foden','Dean Henderson'}

print(England)
```

#Dictionary
Dictionary adalah tipe data pada python yang berfungsi untuk menyimpan kumpulan data/nilai dengan pendekatan “key-value”. As of Python version 3.7, dictionaries are ordered. In Python 3.6 and earlier, dictionaries are unordered.

Source : https://www.w3schools.com/python/python_dictionaries.asp

```python
#Contoh Dictionary 1
product1 = {
  "brand": "Kingston",
  "model": "USB Flash Drive V22",
  "size": "256 GB",
  "year": 2022
}
print(product1["brand"])
```

```python
#Contoh Dictionary 2
Biodata = {
  'nama': 'Robert Langdon',
  'usia': 23,
  'asal': 'Sleman'
}
print(Biodata['usia'])
```

#String Processing
String adalah jenis yang paling populer di bahasa pemrograman. Kita bisa membuatnya hanya dengan melampirkan karakter dalam tanda kutip. Python memperlakukan tanda kutip tunggal sama dengan tanda kutip ganda. Membuat string semudah memberi nilai pada sebuah variabel.

Source : https://jagongoding.com/python/menengah/manipulasi-string-part-1/ 

```python
#Contoh String Concatenation
kata1 = 'Universitas'
kata2 = ' Amikom'
kata3 = ' Yogyakarta'
kalimat=kata1 + kata2 + kata3

print(kalimat)
```

```python
#Contoh String Slicing
Z = 'Rara Mandalika'
X = 'is a Cloud Engineer'

print ("Z[5]:", Z[5])
print ("X[5:10]:", X[5:10])
```

```python
#Contoh String Conversion
kata1 = '18830177'
kata1= int (kata1)+666

print (kata1)
```

```python
#Contoh String Read
nama=input('Nama Anda : ')
usia=input('Usia Anda Tahun Lalu: ')
usia= int(usia) + 1

print ("Name: %s"% nama ,"\t Usia Saat Ini:%d"% usia)
```
