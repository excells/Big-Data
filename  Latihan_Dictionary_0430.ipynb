#SOAL 1

- Buatlah sebuah dictionary berisi 6 mata kuliah seperti tabel berikut:



| Kode Mata Kuliah | Nama Mata Kuliah |
| --- | --- |
| TK004 | Matematika Teknik |
| TK005 | Fotografi |
| TK007 | Instalasi Hardware |
| TK008 | Sistem Operasi |
| TK013 | Pendidikan Pancasila |
| TK015 | Algoritma Komputer dan Struktur Data |

- Buatlah 5 buah **input** berturut-turut untuk memasukkan mata kuliah baru seperti pada tabel berikut:

| Kode Mata Kuliah | Nama Mata Kuliah |
| --- | --- |
| TK011 | Pendidikan Agama |
| TK016 | Jaringan Komputer |
| TK021 | Pemrograman Web |
| TK023 | Multimedia |
| TK040 | Ethical Hacking |

- Ganti kode mata kuliah **TK011** dengan mata kuliah **Pemrograman Python**, dan kode mata kuliah **TK023** dengan mata kuliah **Bahasa Indonesia**

- Cetak dictionary tersebut menggunakan perulangan for sesuai denga format berikut:

```
TK004 ==> Matematika Teknik
TK005 ==> Fotografi
TK007 ==> Instalasi Hardware
TK008 ==> Sistem Operasi
TK013 ==> Pendidikan Pancasila
TK015 ==> Algorima Komputer dan Struktur Data
TK011 ==> Pemrograman Python
TK016 ==> Jaringan Komputer
TK021 ==> Pemrograman Web
TK023 ==> Bahasa Indonesia
TK040 ==> Ethical Hacking
```





````
# uatlah sebuah dictionary berisi 6 mata kuliah seperti tabel berikut:
print("================================== Daftar Mata Kuliah ==================================")
print("#A")
kamus = {'TK004': 'Matematika Teknik', 'TK005': 'Fotografi', 'TK007': 'Instalaso Hardware', 'TK008': 'Sistem Operasi', 'TK013': 'Pendidikan Pancasila', 'TK015': 'Algoritma Komputer dan Struktur Data'}

for key in kamus.keys():
    print(key, "->", kamus[key])
    
#Buatlah 5 buah input berturut-turut untuk memasukkan mata kuliah baru seperti pada tabel berikut:
print("================================== Masukan Mata Kuliah Baru ==================================")
print("#B")
kamus["TK011"] = input(" Masukan nama mata kuliah: ")
kamus["TK016"] = input(" Masukan nama mata kuliah: ")
kamus["TK021"] = input(" Masukan nama mata kuliah: ")
kamus["TK023"] = input(" Masukan nama mata kuliah: ")
kamus["TK040"] = input(" Masukan nama mata kuliah: ")
print("================================== Daftar Mata Kuliah ==================================")
for key in kamus.keys():
    print(key, "->", kamus[key])
    
# Ganti kode dan mata kuliah
print("================================== Daftar Mata Kuliah ==================================")
print("#C")
kamus["TK011"] = "Pemograman Python"
kamus["TK023"] = "Bahasa Indonesia"

for key in kamus.keys():
    print(key, "==>", kamus[key])

````

#SOAL 2
- Buatlah sebuah **dictionary** untuk menampung biodata salah satu teman sekelah anda (seperti **nim, nama, alamat, umur, hobi**), kemudian lakukan hal berikut:
1. Buatlah kode untuk menampilkan key dari dictionary tersebut
2. Buatlah kode untuk menambahkan key berupa **nilai_tugas**
3. Isi **nilai_tugas** tersebut dengan **5 buah nilai** yang tersimpan dalam list
4. Buatlah kode untuk menghitung nilai rata-rata **nilai_tugas**


````
#soal 2 jawab disini:
# Membuat Dictionary
print("======================================================")
print("#A")
biodata = {
    "nim": "19.83.0430",
    "nama": "Fadzillah Excel Aditya",
    "alamat": "Jepang",
    "umur": 18,
    "hobi": "Bermain",
    "nilai_tugas": {
        "b_jepang": 100,
        "Matematika": 90,
        "Pendidikan Pancasila": 100,
        "Fotografi": 90,
        "b_jerman": 85,
    } 
}
for key in biodata.keys():
    print(key, "==", biodata[key])

print()

# Mengakses isi dictionary
print("======================================================")
print("#B")
print("%s" % biodata["nama"])
print("Nilai Bahasa Jepang: %s" % biodata["nilai_tugas"]["b_jepang"])
print("Nilai Matematika: %s" % biodata["nilai_tugas"]["Matematika"])
print("Nilai Pendidikan Pancasila: %s" % biodata["nilai_tugas"]["Pendidikan Pancasila"])
print("Nilai Fotografi: %s" % biodata["nilai_tugas"]["Fotografi"])
print("Nilai Bahasa jerman: %s" % biodata["nilai_tugas"]["b_jerman"])

print()

# nilai rata-rata
print("======================================================")
print("#C")
list_nilai = [100,90,100,90,85]
jumlah_nilai = sum(list_nilai)
jumlah_elemen = len(list_nilai)
nilai_rata_rata = jumlah_nilai / jumlah_elemen
print('nilai: ', list_nilai)
print('rata-rata nilai pada list: ',nilai_rata_rata)
````
