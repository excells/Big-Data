```
# List dengan nama personil BTS
bts = ["Kim Tae-Hyung", "Min Yoon-Gi","Jung Ho-Seak"]
print(bts)
# Hapus Jung Ho-Seak
bts.remove("Jung Ho-Seak")
print(bts)
# Menambah list Personil Bts menggunkan fungsi input
bts.append("Park Ji-Min")
bts.append("Kim Seok-Jin")
bts.append("Kim Nam")
print(bts)
#Update Personil Park Ji-MIn menjadi Jeong-Guk
bts[2] = "Jeong-Guk"
print(bts)
#Hapus Kim Nam
bts.remove("Kim Nam")
print(bts)
#Menampilkan list menggunakan For
for i in range(1):
    print("Nama Personil Bts :" ,bts)
#Menampilkan List menggunakan while
i = 0
while i < len(bts):
    print("Nama Personil dari Bts :", bts)
    i += 1
```

Matrix Perkalian

```
mat1 = [
    [7, 8],
    [9, 10],
]

mat2 = [
    [7, 8],
    [9, 10],
]

mat3 = []

for i in range(0, len(mat1)):
    row = []
    for x in range(0, len(mat1[0])):
        total = 0
        for z in range(0, len(mat1)):
            total = total + (mat1[i][z] * mat2[z][x])
        row.append(total)
    mat3.append(row)

for i in range(0, len(mat3)):
    for x in range(0, len(mat3[0])):
        print (mat3[i][x], end=' ')
    print ()
```
